```python
# Lista vacía
recepcionista = []

print("===================================")
print(" 🤖 RECEPCIONISTA IA ")
print("===================================")

nombre_usuario = input("Ingresa tu nombre: ")
print("Hola", nombre_usuario, "😊")

while True:
    print("\n----- MENÚ -----")
    print("1. Registrar cliente")
    print("2. Ver lista de clientes")
    print("3. Eliminar último cliente")
    print("4. Salir")

    opcion = int(input("Elige una opción: "))

    if opcion == 1:
        print("\n📋 Nuevo cliente")

        nombre = input("Nombre del cliente: ")
        edad = input("Edad: ")
        motivo = input("Motivo: ")

        # Guardamos todo junto en la lista
        registro = nombre + " | " + edad + " años | " + motivo

        recepcionista = recepcionista + [registro]

        print("✔ Cliente registrado\n")

    elif opcion == 2:
        print("\n📋 LISTA DE CLIENTES:")

        if recepcionista == []:
            print("No hay registros\n")
        else:
            i = 0
            while i < len(recepcionista):
                print("Cliente", i + 1, ":", recepcionista[i])
                i = i + 1
            print()

    elif opcion == 3:
        if recepcionista == []:
            print("❌ No hay clientes para eliminar\n")
        else:
            eliminado = recepcionista[-1]
            recepcionista = recepcionista[:-1]
            print("🗑 Se eliminó:", eliminado, "\n")

    elif opcion == 4:
        print("👋 Gracias por usar la Recepción IA,", nombre_usuario)
        break

    else:
        print("⚠ Opción no válida\n")
```
