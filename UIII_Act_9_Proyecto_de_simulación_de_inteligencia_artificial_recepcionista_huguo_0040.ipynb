```python
# --- SIMULADOR DE RECEPCIONISTA IA CON LISTA ORDENADA ---

# 1. Inicialización de la lista y el estado del sistema
registro_huespedes = []
sistema_activo = True

print("--- BIENVENIDO AL SISTEMA DE RECEPCIÓN INTELIGENTE ---")

# 2. Ciclo principal de interacción (IA)
while sistema_activo:
    print("\n[IA]: Hola, elija una opción del menú:")
    print("1. Registrar Huésped")
    print("2. Ver Lista Ordenada de Huéspedes") # Esta opción ahora ordena los datos
    print("3. Calcular Presupuesto")
    print("4. Salir")

    opcion = input("Seleccione (1-4): ")

    # 3. Registro de nuevos datos
    if opcion == "1":
        nuevo_nombre = input("Ingrese el nombre del huésped: ")
        registro_huespedes.append(nuevo_nombre) # Agrega el elemento al final de la lista [4]
        print(f"IA: {nuevo_nombre} ha sido guardado exitosamente.")

    # 4. Mostrar lista con ORDEN ALFABÉTICO
    elif opcion == "2":
        # Primero, la IA ordena la lista internamente de la A a la Z
        registro_huespedes.sort() # Método para ordenar la lista en su lugar [1]

        print("--- LISTA DE HUÉSPEDES (ORDEN ALFABÉTICO) ---")
        indice = 0

        # Verificamos si hay elementos usando el operador relacional de igualdad [5]
        if len(registro_huespedes) == 0:
            print("IA: El registro está vacío.")

        # Recorrido manual de la lista utilizando un ciclo while
        while indice < len(registro_huespedes):
            # Accedemos a cada posición de la lista mediante el índice [4]
            print(f"{indice + 1}. {registro_huespedes[indice]}")
            indice += 1 # Incremento aritmético para pasar al siguiente elemento [6]

    # 5. Lógica de cálculos financieros
    elif opcion == "3":
        noches = int(input("¿Cuántas noches desea cotizar?: "))
        # Operador aritmético de multiplicación para el costo base [7]
        total = noches * 120

        # Operador relacional para aplicar beneficios [8]
        if noches >= 7:
            total = total * 0.85 # Descuento del 15% mediante multiplicación [7]
            print(f"IA: Por estadía larga, el total con descuento es: ${total}")
        else:
            print(f"IA: El costo total de la estancia es: ${total}")

    # 6. Finalización del programa
    elif opcion == "4":
        print("IA: Sesión finalizada. ¡Tenga un buen día!")
        sistema_activo = False # Cambiamos el valor lógico para romper el ciclo while [5]

    # Manejo de entradas incorrectas
    else:
        print("IA: Opción no válida. Por favor, intente de nuevo.")
```
